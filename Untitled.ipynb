```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("final_data_us.csv")
```

```python
df['Total Sales'] = df['Volume Sold (Liters)'] * df['Bottles Sold'] * df['Sale (Dollars)'] * \
                                df['Pack']
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python

df_filtered = df[df['Date'].dt.year >= 2015]  # Filter data from 2015 onwards

df_grouped = df_filtered.groupby([df_filtered['Date'].dt.year, df_filtered['Date'].dt.month]).sum().reset_index()
df_grouped.rename(columns={'Date': 'Year', 'Date': 'Month'}, inplace=True)  # Rename grouped columns
```

```python
df_grouped.shape
```

```python
df_grouped.sort_values(by=['Year', 'Month'], ascending=True, inplace=True)
```

```python
from statsmodels.tsa.stattools import adfuller
#Ho: It is non stationary
#H1: It is stationary

def adfuller_test(sales):
    result=adfuller(sales)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")
```

```python
adfuller_test(df_grouped['Total Sales'])
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing, SimpleExpSmoothing
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
models = []
specifications_category = [
    {
        'name': 'Fixed Parameters',
        'model': ExponentialSmoothing(
            df_grouped['Total Sales'],
            seasonal_periods=12,
            trend='add',
            seasonal='add',
            damped_trend=True
        ).fit(
            smoothing_level=0.2,
            smoothing_trend=0.1,
            smoothing_seasonal=0.1,
            damping_trend=0.98,
            optimized=False
        )
    },
    {
        'name': 'Multiplicative Seasonal',
        'model': ExponentialSmoothing(
            df_grouped['Total Sales'],
            seasonal_periods=12,
            trend='add',
            seasonal='mul',
            damped_trend=True
        ).fit(
            smoothing_level=0.2,
            smoothing_trend=0.1,
            smoothing_seasonal=0.1,
            damping_trend=0.98,
            optimized=False
        )
    },
    {
        'name': 'Multiplicative Trend',
        'model': ExponentialSmoothing(
            df_grouped['Total Sales'],
            seasonal_periods=12,
            trend='mul',
            seasonal='add',
            damped_trend=True
        ).fit(
            smoothing_level=0.2,
            smoothing_trend=0.1,
            smoothing_seasonal=0.1,
            damping_trend=0.98,
            optimized=False
        )
    }
]
```

```python
results = []

for spec in specifications_category:
    model = spec['model']
    name = spec['name']

    # Calculate metrics
    rmse = np.sqrt(mean_squared_error(df_grouped['Total Sales'], model.fittedvalues))
    mae = mean_absolute_error(df_grouped['Total Sales'], model.fittedvalues)

    results.append({
        'name': name,
        'rmse': rmse,
        'mae': mae,
        'model': model
    })
```

```python
best_model = min(results, key=lambda x: x['rmse'])
hw1_model = best_model['model']
hw1_forecast = hw1_model.forecast(12)
```

```python
plt.figure(figsize=(15, 12))

# First subplot: All models
plt.subplot(3, 1, 1)
plt.plot(df_grouped.index, df_grouped['Total Sales'], label='Historical Data', color='blue')

for result in results:
    plt.plot(result['model'].fittedvalues.index,
             result['model'].fittedvalues,
             label=f"{result['name']} Fitted",
             alpha=0.5)

plt.title('Model Comparison')
plt.legend()
plt.grid(True, alpha=0.3)

# Second subplot: Best model forecast
plt.subplot(3, 1, 2)
plt.plot(df_grouped.index,df_grouped['Total Sales'], label='Historical Data', color='blue')
plt.plot(hw1_forecast.index, hw1_forecast,
         label=f'Forecast (Best model: {best_model["name"]})',
         color='green', linestyle='--')

plt.title(f'Best Model Forecast: {best_model["name"]}')
plt.legend()
plt.grid(True, alpha=0.3)

# Third subplot: Residuals of best model
plt.subplot(3, 1, 3)
residuals = df_grouped['Total Sales'] - hw1_model.fittedvalues
plt.plot(df_grouped.index, residuals, label='Residuals', color='red')
plt.axhline(y=0, color='r', linestyle='-', alpha=0.3)
plt.title('Best Model Residuals')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
